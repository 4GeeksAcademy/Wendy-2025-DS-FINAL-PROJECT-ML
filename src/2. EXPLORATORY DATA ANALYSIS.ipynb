# 2.1. CLEANED DATA SET 3 PARTS DIVISION

```python
training_data, temp_data = train_test_split(df_cleaned, test_size=0.4, random_state=42)
testing_data, anonymous_data = train_test_split(temp_data, test_size=0.5, random_state=42)
training_data.to_csv("training_data.csv", index=False)
testing_data.to_csv("testing_data.csv", index=False)
anonymous_data.to_csv("anonymous_data.csv", index=False)
```
